```python
from helpers.extractor import *
from helpers.sound_tools import *
from helpers.neural_network import *

N_STEPS = 32
samplerate = 441*10
compression_rate = 32

loader = SoundLoader(sound="inputs/sounds/piano3.wav",
                     fixed_size=N_STEPS, 
                     sample_size=0, 
                     one_hot = 64,
                     uLaw = 64,
                     amplitude = 1.5,
                     samplerate = samplerate,
                     random = True,
                     cache_size = False,
                     )

network = AutoEncoder(loader,
                      hidden=[16, 4],
                      n_reccurent_input = 0,
                     )

network.index_step = 64

network.n_input = N_STEPS
data = loader.converter.getRawAudio()

features_map = network.EncodeSequence(
                        data, 
                        restore_path="graphs/AbstractNet1/auto_encoder",
                        display_step = 1000, 
                        compression_rate = compression_rate
                        )

result = network.DecodeSequenceWithIndex(
                        features_map,
                        restore_path="graphs/AbstractNet1/auto_encoder",
                        display_step = 10,)

print("input/result")
plt.plot(data)
plt.plot(result, color="red")
plt.show()

print("features")
plt.plot(features_map)
plt.show()
converter = SoundConverter("")
data = converter.TensorToSound(result, "outputs/generated.wav", multiplier=1, offset=0, samplerate=int(loader.samplerate*1))

import IPython
IPython.display.Audio("outputs/generated.wav")
```
